```python
def income(var) :
  if var<5000 : 
   return "low income group"
  elif var>=5000 and var<10000:
   return "medium income group"
  else:
   return "high income group"
```

```python
income(4000)
```

```python
df1["binning"]=df1["totalincome"].apply(income)
```

```python
def sta(var):
    if (var["Education"] == "Graduate") & (var["Married"] == "Yes") & (var["Self_Employed"] == "No"):
        return "stable"
    else:
        return "unstable"
```

```python
df1["stability"]=df1.apply(sta,axis=1)
```
