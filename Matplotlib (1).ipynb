```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Line graph

x=[5,2,5,7,9]
y=[10,5,8,4,5]


plt.figure(figsize=(10,5))
plt.plot(x,y,"o-.k",mec="r")

#labelling x and y axis

plt.xlabel("Profit")
plt.ylabel("Sales")
plt.title("Sales Analysis")

```

```python
help(plt.plot)
```

```python
#stacked plots:

x_1=["A","B","C","D"]
y_1=[10,20,10,30]
y_2=[20,25,15,25]

plt.bar(x_1,y_1,color="k")
x=plt.bar(x_1,y_2,bottom=y_1,color="r")
```

```python
#shapes

var=plt.Circle((0,0),radius=0.5,fc="g")
plt.gca().add_patch(var)
plt.axis("scaled")
```

# Annotation in the graph:

```python
help(plt.annotate)
```

```python
fig,test=plt.subplots()

t=np.arange(0.0,5.0,0.001)
s=np.cos(3*np.pi*t)

line=test.plot(t,s,lw=2)


test.annotate("Local Max",xy=(3.3,1),xytext=(3,1.8),arrowprops=dict(facecolor="red",shrink=0.05))
```
